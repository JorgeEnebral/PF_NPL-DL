```python
from transformers import pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment")
```

```python
sentences = [
    "I love sunny days!",
    "This movie was pretty boring.",
    "I'm feeling fantastic about this new project.",
    "This meal tastes terrible.",
    "It's an average day, nothing special.",
    "This is such a greaaaat movie... 🙄",
    "Tengo Dolor",
]

# Run sentiment analysis
results = sentiment_analyzer(sentences)

# Display the results in a nice format
for sentence, result in zip(sentences, results):
    label = result['label']
    score = result['score']
    print(f"Sentence: '{sentence}'")
    print(f"Sentiment: {label} (confidence: {score:.2f})")
    print('-' * 60)
```

# DATASET CONLL2003

```python
from datasets import load_dataset

# Cargar el dataset CoNLL-2003
dataset = load_dataset("conll2003")

# Ver ejemplos
print(dataset)
print(dataset["train"][0])
```

```python
# Obtener las etiquetas de NER en texto
ner_labels = dataset['train'].features['ner_tags'].feature.names

# Ejemplo de oración
example = dataset['train'][0]

# Decodificar etiquetas de NER
decoded_ner = [ner_labels[tag] for tag in example['ner_tags']]

# Mostrar resultado
for token, ner in zip(example['tokens'], decoded_ner):
    print(f"{token}: {ner}")
```

# MODELO NER PRE-ENTRENADO SPACY

```python
import spacy

# Cargar el modelo de spaCy
nlp = spacy.load("en_core_web_sm")

# Tomar una oración de ejemplo
sentence = " ".join(dataset["train"][0]["tokens"])  # Unir tokens en una sola frase

# Procesar con spaCy
doc = nlp(sentence)

# Mostrar las entidades reconocidas
print(f"Oración: {sentence}\n")
print("Entidades reconocidas por spaCy:")
for ent in doc.ents:
    print(f"{ent.text} → {ent.label_}")
```

```python
entities = nlp.get_pipe("ner").labels

# Mostrar todas las entidades con su explicación
print("Entidades reconocidas por spaCy:")
for entity in entities:
    print(f"{entity} → {spacy.explain(entity)}")
```

# CREAR DATASET NER + SA CON BERT

```python
import pandas as pd

# Cargar modelos de Hugging Face
sentiment_analyzer = pipeline("sentiment-analysis")  # Análisis de sentimiento

# Función para procesar datos y crear DataFrame
def create_dataframe(split):
    sentences = [" ".join(example["tokens"]) for example in dataset[split]]  # Oraciones completas
    ner_tags = [example["ner_tags"] for example in dataset[split]]  # Etiquetas NER

    # Obtener análisis de sentimiento para cada oración
    sentiment_results = [int(sentiment_analyzer(sentence)[0]["label"][-1]) for sentence in sentences]

    # Crear DataFrame
    df = pd.DataFrame({
        "sentence": sentences,
        "ner_tags": ner_tags,
        "sentiment": sentiment_results
    })

    return df

# Crear DataFrames para train, validación y test
df_train = create_dataframe("train")
df_valid = create_dataframe("validation")
df_test = create_dataframe("test")

# Mostrar las primeras filas del dataset de entrenamiento
print(df_train.head())
```
